# Create a Game of Hangman

```python
# Don't need to do graphics
# Use categories
# Show underscores for letters not guessed correctly yet
# Replace underscore with correct letter when guessed correctly.

# Importing for clearing capabilities
from IPython.display import clear_output
import random

# Step 1: Create random categories and words
categories = ["Sports", "Office Supplies"]
words = [["Sports", "cubs"], ["Sports", "giants"], ["Sports", "patriots"],["Office Supplies", "chair"], 
         ["Office Supplies", "desk"], ["Office Supplies", "paperclip"]]

# Step 2a: Create function that writes a list of the correct number of underscores for the word chosen
def displayUList(answer, undList):
    # Turn answer into list
    answerList = list(answer)
    
    if len(undList) == 0:
        # Make the same number of underscores in undList as letters in answerList
        for i in range(len(answerList)):
            undList.append("_ ")
    
    # Print underscore list
    for i in range(len(undList)):
        print(undList[i], end="")
    return undList

# Display categories
def displayCategories():
    counter = 1
    for category in categories:
        print("{}: {}".format(counter, category))
        counter = counter + 1
        
# Get random value from lists based on correct category
def pickRandomWord(word):
    categoryChosen = ""
    rndList = []
    
    # Match word variable with correct category
    for category in categories:
        if category == word:
            categoryChosen = category

    # Loop through words array and fill rndList list with all values that have correct category    
    for i in range(len(words)):
        for j in range(2):
            if categoryChosen == words[i][j]:
                rndList.append(words[i][j + 1])
    
    # Choose random value from list
    return random.choice(rndList)

# Change correct guessed letters
def changeUList(undList, guess, i):
    undList[i] = guess
    return undList
    
# Check to see if guess is in word
def checkGuess(guess, answer, undList):
    # Turn answer into list
    answerList = list(answer)
    # check number of similar letters
    similar = 0
    for i in range(len(answerList)):
        if guess.lower() == answerList[i].lower():
            similar += 1
    
    flag = False
    
    for i in range(len(answerList)):
        if guess.lower() == answerList[i].lower() and similar > 0:
            # Change guess HERE so it changes all same letters in word
            changeUList(undList, guess, i)
            similar -= 1
            flag = True
    return flag
    
# If player guesses word change flag to true
def checkGameOver(answer, flag, undList):
    if list(answer) == undList:
        clear_output()
        print("You win!")
        flag = True
    else:
        flag = False
        
    return flag
    
while True:
    # Clear all output from before
    clear_output()
    undList = []
    lives = 7
    flag = False
    
    # Step 2b: Ask user to pick a category
    displayCategories()
    word = input("Which category would you like to guess from: ")
    clear_output()
    
    # Step 2c: Pick random value from dictionary with correct key
    answer = pickRandomWord(word)

    # Step 2d: Display correct amount of underscores for length of word
    displayUList(answer, undList)
    
    while True:
        # Step 3: Ask for user input to guess letter
        print("")
        guess = input("Guess a letter: ")
        clear_output()
        
        # Step 4: Check if letter is in word
        livesFlag = checkGuess(guess, answer, undList)
        if livesFlag == False:
            lives = lives - 1
            
        # Print number of lives left
        print("{} lives left".format(lives))
        
        # Step 5: Either replace underscore with correct letter or reduce lives by 1
        displayUList(answer, undList)
        flag = checkGameOver(answer, flag, undList)

        # Step 6: Loop until lives = 0 or word is guessed correctly
        if flag == True or lives <= 0:
            if lives <=0:
                print("You ran out of lives! Sorry you lost...")
            break
        
    # Step 7: Ask if they would like to play again
    ans = input("Would you like to play again? ")
    if ans.lower() == "no":
        print("Thanks for playing!")
        break
```
